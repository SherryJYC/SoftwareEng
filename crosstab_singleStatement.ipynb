```python
s1_tab = pd.crosstab(index=coverage['s1'],columns=coverage['r'])
s1_tab.index = ['uncovered','covered']
s1_tab.columns = ['passed','failed']
s1_tab
```

```python
s1_tab = pd.crosstab(index=coverage['s1'],columns=coverage['r'],margins=True)
s1_tab.index = ['uncovered','covered','coltotal']
s1_tab.columns = ['passed','failed','rowtotal']
s1_tab
```

```python
type(s1_tab)
```

```python
stats, p, dof, expected = chi2(s1_tab)
stats,p
```

chi2 statistic

```python
N = s1_tab.iloc[2]['rowtotal']
m = stats/N
m
```

contingency coefficient

```python
proportion = s1_tab.iloc[1]['failed']/s1_tab.iloc[2]['failed']/(s1_tab.iloc[1]['passed']/s1_tab.iloc[2]['passed'])
proportion
```

The statistic to define the failed or the successsful execution result that is more associated with the coverage.

```python
if proportion==1:
    final = 0
elif proportion<1:
     final = -m
else: final =m
final
```

Final statistic for ranking. The higher the value, the higher the suspicious level.
